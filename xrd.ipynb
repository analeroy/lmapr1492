```python
#imports 
from pymatgen import Structure
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from jupyter_jsmol import JsmolView
from IPython.display import display
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
#création structure 
with MPRester("4cITfHbiVnRj5dfqR") as m:    
    struct = m.get_structure_by_material_id("mp-4636")
    w = CifWriter(struct, 0.1)
    w.write_file('mystructure.cif')
```

```python
#Création d'un XRDCalculator pour nous permettre de calculer les 3 premiers pics du diffractogramme correspondant pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.
xrd = XRDCalculator(1.54060)
#Diffractogramme
diff = xrd.get_pattern(struct)
```

```python
#Premier pic du diffractogramme
#Indices hkl
ind1 = diff.hkls[0]
#Angles
ang1 = diff.x[0]
#Intensité
int1 = diff.y[0]
#Résultats
print("Résultats")
print("Les indices hkl associés au premier pic de diffraction sont {}".format(ind1))
print("L'angle du premier pic de diffraction est : 2θ = {} ° ".format(ang1))
print("L'intensité du premier pic de diffraction est {} ".format(int1))
```

```python
#Deuxième pic du diffractogramme
#Indices hkl
ind2 = diff.hkls[1]
#Angles
ang2 = diff.x[1]
#Intensité
int2 = diff.y[1]
#Résultats
print("Résultats :")
print("Les indices hkl associés au deuxième pic de diffraction sont {}".format(ind2))
print("L'angle du deuxième pic de diffraction est : 2θ = {} ° ".format(ang2))
print("L'intensité du deuxième pic de diffraction est {} ".format(int2))
```

```python
#Troisième pic du diffractogramme
#Indices hkl
ind3 = diff.hkls[2]
#Angles
ang3 = diff.x[2]
#Intensité
int3 = diff.y[2]
#Résultats
print("Résultats :")
print("Les indices hkl associés au troisième pic de diffraction sont {}".format(ind3))
print("L'angle du troisième pic de diffraction est : 2θ = {} ° ".format(ang3))
print("L'intensité du troisième pic de diffraction est {} ".format(int3))
```

```python
#Plot du diagramme de diffraction
#Visualisation des 3 premiers pics du diffractogramme correspondant pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å
xrd.get_plot(struct, two_theta_range = (15, 33), annotate_peaks = True, ax = None, with_labels = True, fontsize = 16)
```

On peut donc vérifier à l'aide de ce difractogramme que nos valeurs trouvées ci-dessus sont bonnes.
